```python
#Rider performance by Gender and Age based on avg trip distance (station to station), median speed (distance traveled / trip duration)
citi_01_2020.head()
```

```python
new_d = citi_01_2020
```

```python
import geopy
from geopy.distance import vincenty 
new_d['Start Coordinates'] = list(zip(new_d['start station latitude'],new_d['start station longitude']))
new_d['End Coordinates'] = list(zip(new_d['end station latitude'],new_d['end station longitude']))
dist = []
for i in range(len(new_d)):
    dist.append(geopy.distance.vincenty(new_d.iloc[i]['Start Coordinates'],new_d.iloc[i]['End Coordinates']).miles)
```

```python
new_d['distance']= dist
```

```python
new_d.dropna()
```

```python
new_d.head()
```

```python
age_d=[]
for i in range(len(new_d)):
    age_d.append(2020 - new_d.iloc[i]['birth year'])
```

```python
#median Speed
med_speed=[]
for i in range(len(new_d)):
    med_speed.append(((new_d.iloc[i]['distance'])/(new_d.iloc[i]['tripduration']))*60)
```

```python
med_speed
```

```python
new_d['med_speed(mil/hr)']=med_speed
new_d['age']=age_d
```

```python
new_d.head()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize = (15,10))
g = sns.lineplot(data = new_d, x='age',y='med_speed(mil/hr)',hue='gender')
```
